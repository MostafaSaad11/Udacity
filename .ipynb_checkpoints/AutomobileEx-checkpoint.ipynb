```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
import seaborn as sns
import matplotlib.pyplot as plt
```

#Data Cleaning
1- replace the ? by NaN
2- convert columns data types (stroke  ,horsepower , peak-rpm ,price )
3- replace - by _ in column name
4- replace the missing values with mean
5- categorieze the price for 3 levels

```python
df= df.replace('?' , np.nan)
```

```python
df.columns = [i.replace('-', '_') for i in df.columns]
```

```python
df.head()
```

```python
# 2- Convert Datatype
df[['stroke' ,'bore' ,'horsepower' , 'peak_rpm' ,'price']] = df[['stroke','bore','horsepower' , 'peak_rpm' ,'price']].apply(pd.to_numeric)
df.info()
```

```python
# 4-replace by mean
def replace_mean(datafrm,col):
    for i in col:
        datafrm[i] = datafrm[i].replace(np.nan , datafrm[i].mean())
        #print (datafrm[i].mean())
```

```python
replace_mean(df,['bore', 'stroke','horsepower' , 'peak_rpm', 'price'])
df.head()
```

```python
df['drive_wheels'].value_counts()
```

```python
def get_uniue_val (datafr): 
    datafr =datafr.select_dtypes(exclude=['float','int64'])
    column_list =  datafr
    for i in column_list :
        #print(i)
        print('Column ' ,i , datafr[i].unique())
```

```python
get_uniue_val(df)
```

```python
def get_columns_strg (datafr): 
    datafr =datafr.select_dtypes(exclude=['float','int64'])
    column_list =  datafr
    for i in column_list :
         if '?' in  datafr[i].unique():
                print(i)
        #print('Column ' ,i , datafr[i].unique())
```

```python
get_columns_strg(df)
```

```python
#5- categorieze the price for 3 levels
price_ranges = (max(df['price'])-min(df['price']))/3
price_cat = range( int(min(df['price'])) , int(max(df['price'])) , int(price_ranges))
df['price_levels'] = pd.cut(df['price'] ,price_cat , labels=['Low','Medium','High' ] )
df['price_levels']
```

```python
print(df['price_levels'].value_counts())
```

```python
print(df['drive_wheels'].value_counts())
```

```python
sns.boxplot(x='drive_wheels' , y='price' , data=df)
```

```python
df_for_piv = df[['drive_wheels' , 'make' , 'price' , 'body_style', 'price_levels']]
pivt_1 = df_for_piv.pivot_table(index= ['drive_wheels','price_levels'], columns=['body_style'] ,values='price', aggfunc='mean', fill_value=0,sort=True).astype(int)
pivt_1 
```

```python
pivt_2 = df_for_piv.pivot_table(index= ['drive_wheels','price_levels'], columns=['make'] ,values='price', aggfunc='mean', fill_value=0,sort=True).astype(int)
pivt_2 
```

```python
pivt_3 = df_for_piv.pivot_table(index= ['drive_wheels'], columns=['body_style'] ,values='price', aggfunc='mean', fill_value=0,sort=True).astype(int)
fig, ax = plt.subplots(figsize=(15,7))  
sns.heatmap(pivt_3 ,ax=ax , cmap="YlGnBu")
```

```python
pivt_4 = df_for_piv.pivot_table(index= ['drive_wheels'], columns=['make'] ,values='price', aggfunc='mean', fill_value=0,sort=True).astype(int)
fig, ax = plt.subplots(figsize=(15,7))  
sns.heatmap(pivt_4 ,ax=ax, cmap="Greens")
```

```python
df_grp = df[['drive_wheels' , 'make' , 'price' , 'body_style']]
x=df_grp.groupby(['drive_wheels' , 'make'  , 'body_style' ]).mean().astype(int)
x["price"] = x["price"].map('{:,d}'.format)
x
```
